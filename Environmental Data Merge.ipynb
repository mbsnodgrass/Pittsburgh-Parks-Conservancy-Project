```python
import pandas as pd

#Import tree canopy and pollution data
tree_parks = pd.read_csv('TreeCanopyPARKS.csv')
pollution_parks = pd.read_csv('PollutionParks.csv')
tree_walkshed = pd.read_csv('TreeCanopyWALKSHEDS.csv')
pollution_walkshed = pd.read_csv('PollutionWALKSHEDS.csv')

#Import sewrshed data
sewer_parks = pd.read_csv('parks_sewer_scores.csv')
sewer_walksheds = pd.read_csv('walkshed_sewer_scores.csv')
```

```python
#Show length of each park dataset

datasets = [tree_parks, pollution_parks, tree_walkshed, pollution_walkshed, sewer_parks, sewer_walksheds]

# Loop through each dataframe and print its length
for i, datasets in enumerate(datasets, start=1):
    print(f"Length of dataset {i}: {len(datasets)}")
```

```python
#Drop all irrelevant columns in all of our datasets

tree_parks_grouped = tree_parks[['objectid','avg_mean']]
```

```python
pollution_parks_grouped = pollution_parks[['objectid','avg_mean']]
```

```python
tree_walkshed_grouped = tree_walkshed[['SiteID','avg_mean']]
```

```python
pollution_walkshed_grouped = pollution_walkshed[['SiteID','avg_mean']]
```

```python
sewer_parks_grouped = sewer_parks[['objectid.x','avg_sewershed_score']]
```

```python
sewer_walksheds_grouped = sewer_walksheds[['SiteID','avg_sewershed_score_walkshed']]
```

```python
#Rename columns for tree canopy and pollution

tree_parks_grouped = tree_parks_grouped.rename(columns={'avg_mean': 'Tree_Canopy_Park'})
pollution_parks_grouped = pollution_parks.rename(columns={'avg_mean': 'Pollution_Park'})
tree_walkshed_grouped = tree_walkshed_grouped.rename(columns={'avg_mean': 'Tree_Canopy_Walkshed'})
pollution_walkshed_grouped = pollution_walkshed_grouped.rename(columns={'avg_mean': 'Pollution_Walkshed'})
```

```python
#Rename Columns for Sewershed

sewer_parks_grouped = sewer_parks_grouped.rename(columns={'avg_sewershed_score': 'Sewershed_Park'})
sewer_walksheds_grouped = sewer_walksheds_grouped.rename(columns={'avg_sewershed_score_walkshed': 'Sewershed_Walkshed'})
```

```python
#Rename walkshed ID's to match Park ID column name
tree_walkshed_grouped = tree_walkshed_grouped.rename(columns={'SiteID': 'objectid'})
pollution_walkshed_grouped = pollution_walkshed_grouped.rename(columns={'SiteID': 'objectid'})
sewer_walksheds_grouped = sewer_walksheds_grouped.rename(columns={'SiteID': 'objectid'})
```

```python
#Rename sewer park ID to match Park ID column name
sewer_parks_grouped = sewer_parks_grouped.rename(columns={'objectid.x': 'objectid'})
```

```python
#The big merge!

environmental = tree_parks_grouped.merge(pollution_parks_grouped, on='objectid').merge(tree_walkshed_grouped, on='objectid').merge(pollution_walkshed_grouped, on='objectid').merge(sewer_parks_grouped, on='objectid').merge(sewer_walksheds_grouped, on='objectid')
environmental.columns
```

```python
#Keep our relevant variables

variables = ['objectid','updatepknm','Tree_Canopy_Park','Tree_Canopy_Walkshed','Pollution_Park','Pollution_Walkshed','Sewershed_Park','Sewershed_Walkshed']

environmental = environmental[variables]
```
